```python
import seaborn as sns
import matplotlib.pyplot as plt

for i in db['Region'].unique():
    region_data = db[db['Region'] == i]
    plt.figure(figsize=(10, 7))

    # Create a KDE plot for the 'Estimated Unemployment Rate(%)' column with integer year values on the x-axis
    sns.kdeplot(region_data['Estimated_Unemployment_Rate_(%)'], shade=True)

    plt.title(f'KDE Plot for Region {i}')
    plt.ylabel('Density')
    plt.xlabel('Estimated Unemployment Rate(%)')
    plt.grid(True)  # Add gridlines
    plt.show()
```

```python
count_by_region=db['Region'].value_counts().rename_axis('State').reset_index(name='Count')
count_by_region.style.background_gradient(cmap='Blues')
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(db['Estimated_Employed'], db['Estimated_Labour_Participation_Rate_(%)'], alpha=0.5)
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.title('Estimated Employed vs. Estimated Labour Participation Rate')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.title('Unemployment Rate based on Area')
sns.scatterplot(y=db['Region'],x=db['Estimated_Unemployment_Rate_(%)'],hue=db['Area'])
```

```python
# Filter data where Estimated_Unemployment_Rate_(%) is greater than 50%
filtered_data = db[db['Estimated_Unemployment_Rate_(%)'] > 50]

# Count the occurrences of each region
region_counts = filtered_data['Region'].value_counts()

# Create the bar graph with custom styling
plt.figure(figsize=(10, 6))  # Set the figure size for better visibility
bars = plt.bar(region_counts.index, region_counts, color='skyblue')  # Customizing bar color

# Adding data labels above each bar within the graph boundaries
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height(), str(int(bar.get_height())), fontsize=12, color='black')

plt.title('Unemployment Rate by Region (Rate > 50%)')
plt.xlabel('Region')
plt.ylabel('Count')
plt.tight_layout()  # Ensures a clean layout
plt.show()
```

```python

# Pairplot to visualize relationships between numerical columns by Area
plt.figure(figsize=(12, 10))
sns.pairplot(data=db, vars=['Estimated_Unemployment_Rate_(%)', 'Estimated_Employed', 'Estimated_Labour_Participation_Rate_(%)'], hue='Area')
plt.suptitle('Pairplot of Numerical Columns by Area', y=1.02)
plt.tight_layout()
plt.show()
```

# Unemployment_Rate_upto_11_2020.csv Analysis

```python
db_2 = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
db_2
```

```python
db_2.info()
```

```python
db_2[' Date'] = db_2[' Date'].astype('datetime64')
```

```python
db_2['year'] = db_2[' Date'].dt.isocalendar().year
db_2['month'] = db_2[' Date'].dt.month
```

```python
import plotly.express as px
unemployment = db_2[['Region', 'Region.1', ' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']]
unemployment = unemployment.groupby(['Region.1', 'Region'])[' Estimated Unemployment Rate (%)'].mean().reset_index()
fig = px.sunburst(unemployment, path=['Region.1', 'Region'], values=' Estimated Unemployment Rate (%)', title='Unemployment rate in each State and Region',height=850)
fig.show()
```

```python
import calendar
import plotly.express as px

# Convert month to integer and get month abbreviation
db_2['Month_int'] = db_2['month'].apply(lambda x: int(x))
db_2['Month_name'] = db_2['Month_int'].apply(lambda x: calendar.month_abbr[x])

# Create the animated bar plot
fig = px.bar(
    db_2,
    x='Region',
    y=' Estimated Unemployment Rate (%)',
    animation_frame='Month_name',
    color='Region',
    title='Unemployment Rate Across Regions (Jan. 2020 - Oct. 2020)',
    height=600,
    width=1000,
    template='plotly_dark',  # Dark background template
    labels=[' Estimated Unemployment Rate (%)'],  # Rename y-axis label
)

# Customizing the layout
fig.update_layout(
    xaxis_title='Region',  # X-axis label
    yaxis_title=' Estimated Unemployment Rate (%)',  # Y-axis label
    xaxis={'categoryorder': 'total descending'},
    uniformtext_minsize=8,  # Minimum font size for text labels
    uniformtext_mode='hide',  # Hide text labels if too many
)

# Customizing the color scale
fig.update_traces(marker_line_color='white', marker_line_width=0.5, opacity=0.8)

# Set animation duration
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 1000  # Faster animation

# Show the improved plot
fig.show()
```
